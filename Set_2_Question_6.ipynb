```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.filters import prewitt, roberts, sobel, laplace
from skimage.feature import canny
from skimage import img_as_float
```

```python
img = cv2.imread("dog.png", cv2.IMREAD_GRAYSCALE)
img = img_as_float(img)
```

```python
# 1. Sobel Edge Detection
edges_sobel = sobel(img)

# 2. Prewitt Edge Detection
edges_prewitt = prewitt(img)

# 3. Roberts Edge Detection
edges_roberts = roberts(img)

# 4. Laplacian of Gaussian (LoG)
blurred = cv2.GaussianBlur(img, (3,3), 0)   # smoothing before Laplacian
edges_log = laplace(blurred)

# 5. Canny Edge Detection
edges_canny = canny(img, sigma=1)
```

```python
# Plot results
titles = ['Original', 'Sobel', 'Prewitt', 'Roberts', 'LoG', 'Canny']
images = [img, edges_sobel, edges_prewitt, edges_roberts, edges_log, edges_canny]

plt.figure(figsize=(12, 8))
for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```
