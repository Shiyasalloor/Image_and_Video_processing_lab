```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import files
```

```python
# Upload two images
print("Upload two images (same size preferred):")
uploaded = files.upload()
file_names = list(uploaded.keys())
```

```python
if len(file_names) < 2:
    print("Please upload two images.")
else:
    image1 = cv2.imread(file_names[0])
    image2 = cv2.imread(file_names[1])

    # Resize second image to match the first if needed
    if image1.shape != image2.shape:
        image2 = cv2.resize(image2, (image1.shape[1], image1.shape[0]))

    # Perform arithmetic operations
    added = cv2.add(image1, image2)
    subtracted = cv2.subtract(image1, image2)
    multiplied = cv2.multiply(image1, image2)
    divided = cv2.divide(image1, image2)

    # Display results
    def show(title, img):
        print(title)
        cv2_imshow(img)

    show("Image 1", image1)
    show("Image 2", image2)
    show("Addition", added)
    show("Subtraction", subtracted)
    show("Multiplication", multiplied)
    show("Division", divided)
```
