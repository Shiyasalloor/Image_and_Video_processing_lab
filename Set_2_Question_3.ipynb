```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import files

# Upload image
print("Upload a image:")
uploaded = files.upload()
image_path = next(iter(uploaded))
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
img = cv2.resize(img, (256, 256))  # Optional resizing for speed
```

```python
def mean_filter(img, ksize=3):
    return cv2.blur(img, (ksize, ksize))

def weighted_mean_filter(img):
    kernel = np.array([[1, 2, 1],
                       [2, 4, 2],
                       [1, 2, 1]], dtype=np.float32) / 16
    return cv2.filter2D(img, -1, kernel)

def gaussian_filter(img, ksize=3, sigma=1):
    return cv2.GaussianBlur(img, (ksize, ksize), sigma)

def laplacian_filter(img):
    return cv2.Laplacian(img, cv2.CV_64F, ksize=3).astype(np.uint8)

def sobel_filter(img, direction='x'):
    if direction == 'x':
        return cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3).astype(np.uint8)
    else:
        return cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3).astype(np.uint8)

def sharpening_filter(img):
    kernel = np.array([[0, -1, 0],
                       [-1, 5,-1],
                       [0, -1, 0]])
    return cv2.filter2D(img, -1, kernel)
```

```python
def show(title, img):
    print(title)
    cv2_imshow(img)

print("Original Image:")
show("Original", img)

# Apply filters
show("1. Mean Filter", mean_filter(img))
show("2. Weighted Mean Filter", weighted_mean_filter(img))
show("3. Gaussian Filter", gaussian_filter(img))
show("4. Laplacian Filter", laplacian_filter(img))
show("5. Sobel Filter (X-direction)", sobel_filter(img, 'x'))
show("6. Sobel Filter (Y-direction)", sobel_filter(img, 'y'))
show("7. Sharpening Filter", sharpening_filter(img))
```
